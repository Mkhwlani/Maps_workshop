## Part 4: Geospatial Map

Finally, let's plot every neighborhood on a map. Bubble **size** = sale price, bubble **color** = gross rent. This uses `hvplot.points()` with `geo=True` (powered by GeoViews).

```python
# Pre-built: load coordinates and merge with neighborhood averages
coords_df = pd.read_csv(
    Path("./Resources/neighborhoods_coordinates.csv"),
    index_col='Neighborhood'
)

neighborhood_avg = sfo_data_df.groupby('neighborhood').mean()

all_neighborhoods_df = (
    pd.concat([coords_df, neighborhood_avg], axis='columns')
    .reset_index()
    .dropna()
    .rename(columns={'index': 'Neighborhood'})
)

all_neighborhoods_df.head()
```

**What to look for:** Large dark bubbles = high sale price AND high rent. Small dark bubbles = affordable to buy, high rental returns — the sweet spot for investment.
